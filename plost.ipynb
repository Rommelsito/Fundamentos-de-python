```python
#Agregar categoria de edads: Infante, Adolescente, Adulto, Adulto
# NiÃ±os < 12, Adolescentes 12-17, Adultos > 18-55, Adulto mayor 56
df_1 = df_titanic[['Name', 'Age', "Sex"]]
```

```python
df_clean = df_1[df_1['Age'].notna()]
```

```python
df_clean['Category'] = df_clean['Age']
```

```python
#Contruir funcion para definir categoria 
# NiÃ±os < 12, Adolescentes 12-17, Adultos > 18-55, Adulto mayor 56

def category(Age):
    if Age <= 12:
        return 'child'
    elif Age > 12 and Age <= 17:
        return 'adolescent'
    elif Age > 18 and Age <= 55:
        return 'Adulto'
    else:
        return 'Adulto mayor'
```

```python
category(5)
```

```python
df_clean['Category'] = df_clean['Category'].apply(category)
```

##### .groupby(). Me permite agrupar categorias o agrupar datos categoricos

```python
data = {
    "category": ["A", "B","A","A","B","C","C","C"],
    "values" : [10, 20, 12, 30, 40, 50, 60, 70],
    "Quantity" : [1, 2, 3, 4, 5, 6, 7, 8]
}
```

```python
df_101 = pd.DataFrame(data)
print(df_101)
```

```python
#Suma de los valores por categoria
df_101.groupby("category")["values"].sum()
```
